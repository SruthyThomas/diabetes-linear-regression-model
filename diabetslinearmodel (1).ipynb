## Building a Linear Regression Model using diabetes dataset(inbuilt dataset in sklearn)

## import libraries

```python

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## import dataset 

```python
from sklearn import datasets
```

```python
diabetes=datasets.load_diabetes()
```

```python
diabetes
```

## Covert dataset to a meaningful dataframe

```python
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df
```

## Adding target to our dataframe

```python
df["Target"]=diabetes.target
df
```

```python
print(diabetes.DESCR)
```

```python
df.info()
```

```python
#statistical summary of dataset
df.describe()
```

## Visualize the dataset

```python
df.hist(bins=50,figsize=(12,10))
```

```python
df.corr()
```

```python
## Splitting the dataset
X=df.iloc[:,:-1]
```

```python
X
```

```python
Y=df.iloc[:,-1]
```

```python
Y
```

## splitting the datset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8)
```

```python
X_train.shape,Y_train.shape
```

```python
X_test.shape,Y_test.shape
```

## important libraries for model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
model_lr=LinearRegression()
model_lr.fit(X_train,Y_train)
```

## Apply trained model to make prediction (on test set)

```python
Y_predict=model_lr.predict(X_test)
```

## print intercept and coefficients

```python
print(model_lr.coef_)
```

```python
print(model_lr.intercept_)
```

## print model performance

```python
mse=mean_squared_error(Y_test,Y_predict)
mse
```

```python
r2score=r2_score(Y_test,Y_predict)
r2score
```

```python

adjustedr2score=1-(1-r2score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```
